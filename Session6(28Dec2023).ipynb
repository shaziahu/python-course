# more on nested ifs

```python
item  = input('What do you wana buy?').lower()
if item=='shirt':
    
    brand  = input('Which brand of the shirt?').lower()
    if brand=='dinner':
        
        color = input("Which color of the shirt?").lower()
        if color=='black':
            
            size  = input("Which size fits you?").lower()
            if size =='medium':
                
                price = float(input("What price do suggest"))
                if price <2000:
                    print("Please pack")
                else:
                    print("Sorry we dont have much budget")
            else:
                print(f"Sorry we dont have {size}")
            
        else:
            print(f"Sorry we dont have {color}")
            
        
    else:
        print(f'Sorry we dont have {brand}')
    
else:
    print(f"Sorry we dont deal in {item}")
    
```
